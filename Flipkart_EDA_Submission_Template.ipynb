## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a new column time difference to find the time taken to respond
# Convert columns to datetime with the correct format
df['Issue_reported at'] = pd.to_datetime(df['Issue_reported at'], format='%d/%m/%Y %H:%M', errors='coerce')
df['issue_responded'] = pd.to_datetime(df['issue_responded'], format='%d/%m/%Y %H:%M', errors='coerce')

# Format the datetime columns to the desired output format (optional for display)
df['Issue_reported_at (formatted)'] = df['Issue_reported at'].dt.strftime('%d/%m/%y')
df['Issue_responded (formatted)'] = df['issue_responded'].dt.strftime('%d/%m/%y')

# Calculate the time difference and convert to minutes
df['Time difference'] = (df['issue_responded'] - df['Issue_reported at']).dt.total_seconds() / 60

# Show first 5 rows
df.head()
````

````python
# Remove columns that have a lot of null values
remove_columns = ['order_date_time', 'Customer Remarks', 'connected_handling_time']

# Check which columns are present in the dataframe
existing_columns_to_remove = [col for col in remove_columns if col in df.columns]

# Remove the columns if they are present
if existing_columns_to_remove :
    df = df.drop(columns = existing_columns_to_remove)
    print(f'Columns {existing_columns_to_remove} have been removed.')
else :
    print('None of the specified columns are present in the data frame or already been removed.')

df.shape
````

### What all manipulations have you done and insights you found?

Columns Having too many null values have been removed which includes customer remarks, connected handling time and order date time. So its been found that CSAT score depends on remaining values. Date format has been changed to appropriate format.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# selecting only columns with numerical data types
numerical_df = df.select_dtypes(include=['number'])

#Finding correlation
corr_matrix = numerical_df.corr()

#Plotting correlation
plt.figure(figsize = (12,6))
sns.heatmap(corr_matrix, annot = True, cmap = 'Set1', fmt = '.2f', linewidths = 0.5)
plt.title('Correlation Heatmap')
# show the plot
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is used because it visually highlights relationships between all numerical features in the dataset, making it easy to spot patterns, strengths, and directions of correlations at a glance.

##### 2. What is/are the insight(s) found from the chart?

This suggests that as price or response time increases, customer satisfaction decreases slightly. However, these correlations are not strong, indicating many other factors may drive satisfaction.

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code

# selecting only columns with numerical data types
numerical_df = df.select_dtypes(include=['number'])

#Plot the pair plot with controlled figure size (approx 4 plots per row)
sns.pairplot(numerical_df, 
             corner=True,
             diag_kind = 'kde', 
             plot_kws = {'alpha' : 0.6, 's':40},
            height = 3, # height of each subplot
            aspect=1) # width = height * aspect ratio
plt.suptitle("Pair Plot", y=1.02, fontsize=26)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a pair plot because it efficiently shows the relationships and distributions among all numerical features in one place, making it easy to spot trends, patterns, and potential correlations.

##### 2. What is/are the insight(s) found from the chart?

The distributions are heavily concentrated near lower values for all features. Most CSAT scores are at discrete points (especially at the top end), and there are no strong linear relationships visible between item price, time difference, and satisfaction scores.
